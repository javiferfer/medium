### Stacked bars

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = [7, 6, 1]
y = [4, 3, 8]
z = [5, 8, 6]
labels = ['Date 1', 'Date 2', 'Date 3']
```

```python
fig, axs = plt.subplots(1, 1, figsize=(10, 6))
pos_bar = np.arange(len(labels))
axs.bar(pos_bar-0.1, x, width=0.2, color='lightgray', align='center', edgecolor='black', label='First variable')
axs.bar(pos_bar-0.1, y, bottom=x, width=0.2, color='lightyellow', align='center', edgecolor='black', label='Second variable')
axs.bar(pos_bar+0.1, z, width=0.2, color='lightblue', align='center', edgecolor='black', label='Third variable')
axs.set_xticks(pos_bar)
axs.set_xticklabels(labels, fontsize=12)
axs.set_ylabel('Value', fontsize=12)
axs.set_title('Stacked bar template', fontsize=15)

plt.legend(fontsize=12)
plt.show()
```

### Boxplots

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from statannot import add_stat_annotation
%matplotlib inline
```

https://github.com/webermarcolivier/statannot/blob/master/example/example.ipynb

```python
np.random.seed(16)
df = pd.DataFrame(np.random.randint(0, 10, size=(90, 1)), columns=['Data'])
df['Labels'] = ['Label_1']*30 + ['Label_2']*30  + ['Label_3']*30
df['Hue'] = ['Hue_1']*10 + ['Hue_2']*10  + ['Hue_3']*10 + ['Hue_1']*10 + ['Hue_2']*10  + ['Hue_3']*10 + ['Hue_1']*10 + ['Hue_2']*10 + ['Hue_3']*10
print(df)
```

```python
x = "Labels"
y = "Data"
hue = "Hue"
hue_order=['Hue_1', 'Hue_2', 'Hue_3']
box_pairs=[
    (("Label_1", "Hue_1"), ("Label_1", "Hue_2")),
    (("Label_1", "Hue_1"), ("Label_2", "Hue_1"))]

fig, axs = plt.subplots(1, 1, figsize=(10, 5))
ax = sns.boxplot(data=df, x=x, y=y, hue=hue, ax=axs)
add_stat_annotation(ax, data=df, x=x, y=y, hue=hue, box_pairs=box_pairs,
                    test='Mann-Whitney', loc='inside', verbose=2)
axs.set_xlabel("", fontsize=14)
axs.legend(loc='upper left', bbox_to_anchor=(1.03, 1))
```

### Lines

```python
from itertools import count
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
var_1 = [0.9, 0.7, 0.6, 0.5, 0.49]
var_2 = [0.95, 0.69, 0.61, 0.30, 0.22]
var_3 = [0.5, 0.49, 0.45, 0.35, 0.33]
var_4 = [0.55, 0.53, 0.52, 0.51, 0.48]
```

```python
fig = plt.figure()
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(15, 6))
value_format = "{:.0f}%"
xticks = ['Input', 'Layer 1', 'Layer 2', 'Layer 3', 'Output']

axes[0].plot(range(len(var_1)), var_1, linestyle='-.', marker='v', markersize=7, linewidth='2', label='Var 1', c='red')
axes[0].plot(range(len(var_2)), var_2, linestyle='dashed', marker='o', markersize=7, linewidth='2', label='Var 2', c='orange')
axes[0].plot(range(len(var_3)), var_3, linestyle='-.', marker='v', markersize=7, linewidth='2', label='Var 3', c='royalblue')
axes[0].plot(range(len(var_4)), var_4, linestyle='dashed', marker='o', markersize=7, linewidth='2', label='Var 4', c='green')
axes[0].set_title('First plot', fontsize=20)
axes[0].set_ylabel('Accuracy', fontsize=14)
axes[0].set_xticks(np.arange(0, len(xticks), 1))
axes[0].set_xticklabels(xticks, fontsize=14)
axes[0].set_yticks(np.arange(0, 1.02, step=0.1))

axes[1].plot(range(len(var_1)), var_1, linestyle='-.', marker='v', markersize=7, linewidth='2', label='Var 1', c='red')
axes[1].plot(range(len(var_2)), var_2, linestyle='dashed', marker='o', markersize=7, linewidth='2', label='Var 2', c='orange')
axes[1].plot(range(len(var_3)), var_3, linestyle='-.', marker='v', markersize=7, linewidth='2', label='Var 3', c='royalblue')
axes[1].plot(range(len(var_4)), var_4, linestyle='dashed', marker='o', markersize=7, linewidth='2', label='Var 4', c='green')
axes[1].set_title('Second plot', fontsize=20)
axes[1].set_ylabel('Accuracy', fontsize=14)
axes[1].set_xticks(np.arange(0, len(xticks), 1))
axes[1].set_xticklabels(xticks, fontsize=14)
axes[1].set_yticks(np.arange(0.4, 1.02, step=0.05))
axes[1].legend(fontsize=15, bbox_to_anchor=(0, -0.1))

def label_subplots(ax_lst, *, upper_case=True, offset_points=(-5, -5)):
    start_ord = 65 if upper_case else 97
    for ax, lab in zip(np.ravel(ax_lst), (chr(j) for j in count(start_ord))):
        ax.annotate(lab, (1, 1),
                    xytext=offset_points,
                    xycoords='axes fraction',
                    textcoords='offset points',
                    ha='right', va='top', fontsize=25)

label_subplots(axes, upper_case=True)
```
